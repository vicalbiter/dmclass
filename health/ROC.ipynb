```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_curve, auc
from scipy import interp
import matplotlib.pyplot as plt
```

```python
def plotROC5Folds(vals, name):
    k = 0
    tprs = []
    aucs = []
    mean_fpr = np.linspace(0, 1, 100)
    for i in range(5):
        scores = vals[:,k]
        labels = vals[:,k+1]
        fpr, tpr, thresholds = roc_curve(labels, scores)
        tprs.append(interp(mean_fpr, fpr, tpr))
        tprs[-1][0] = 0.0
        roc_auc = auc(fpr, tpr)
        aucs.append(roc_auc)
        plt.plot(fpr, tpr, lw=1, alpha=0.3, label="ROC fold %d (AUC=%0.2f)" % (i + 1, roc_auc))
        k = k + 2
    
    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)

    plt.plot(mean_fpr, mean_tpr, lw=2, color='b', alpha=0.8, label='Mean ROC (AUC=%0.2f)' % (mean_auc))
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title(name)
    plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',
           label='Random', alpha=.8)
    plt.legend(loc="lower right")
    plt.show()
```

```python
data = pd.read_csv('rocAllf.csv')
vals = data.values
plotROC5Folds(vals, 'ROC - All Features Model')
```

```python
data = pd.read_csv('roc5f.csv')
vals = data.values
plotROC5Folds(vals, 'ROC - 5 Features Model')
```
